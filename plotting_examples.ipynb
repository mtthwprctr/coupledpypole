```python
import plotting, lattices

square = lattices.Square(1)

plotting.plot_lattice(square)
```

The radius will default to $R_0/3$ (the limit of the coupled dipole approximation), but can be overridden by changing the argument _r_

```python
kagome = lattices.Kagome(1, -0.1)
plotting.plot_lattice(kagome, r=0.15)
```
